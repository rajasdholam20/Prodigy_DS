```python
#checking for total Null values

df.isnull().sum()
```

```python
# checking for the null values & observing there are 177 null values in Age & 687 null values in Cabin
```

```python
# now filling the null values of age with median of age

df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
#now checking if the values are imputed succesfully
a = df['Age'].isnull().sum()
print("The total missing values in 'Age' are : ", a)
```

```python
#the values were imputed succesfully
# now imputing the values of Embarked & Cabin columns with mode
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace = True)
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace = True)
```

```python
b = df['Embarked'].isnull().sum()
print("The total missing values in 'Embarked' are : ", b)
```

```python
c = df['Cabin'].isnull().sum()
print("The total missing values in 'Cabin' are : ", c)
```

Data Visualization

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

#UNIVARIATE ANALYSIS
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Age',kde=True)
plt.title('Histogram to understand Age')
plt.show()
# 'kde' adds smooth curves.
```

The data suggests a bimodal distribution of ages, with two prominent peaks. This could indicate two predominant age groups within the dataset. The kernel density estimate, which smooths out the frequency of occurrences, shows that the most common ages in the dataset are around the mid-20s to 30s and mid-50s to 60s. This type of age distribution might be seen in organizations with a mix of young professionals and experienced individuals, possibly indicating a diverse workforce.

The histogramâ€™s range from 0 to 80 years and the count axis suggest a sizable sample size, allowing for a reliable analysis of the age structure within the population studied. The visualization aids in understanding the age dynamics and could be useful for making decisions related to marketing, policy-making, or resource allocation based on age-related trends.

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Age',kde=True, hue='Sex', multiple = 'dodge')
plt.title('Histogram to understand Age in accordance with Sex')
plt.show()

# 'hue' is used to add another column to the histgram to compare with present column.
# 'multiple' is used to seperate the bars od both column so they do not overlap each other
```

Age Distribution Peaks:
The largest age group for both males and females falls between approximately 20-30 years old.
This peak suggests a significant proportion of young adults in the population.
Normal Distribution:
Both male and female age distributions follow a normal curve.
As age increases beyond the peak range, the counts taper off gradually.
Gender Differences:
Males exhibit a slightly higher count in the 20-30 age range compared to females.
This information could be valuable for targeted marketing or demographic planning.
In summary, the data indicates a youthful population with a normal age distribution, and gender-specific variations exist within the age groups. 

```python
plt.figure(figsize=(26,10))
sns.barplot(data=df, x='Age', y='Survived', hue='Sex',estimator=sum,errorbar=None)
plt.title('Barplot show casing Distribution between Age & Survived in accordance to Sex')
plt.show()
```

```python
#This is the optional we can take to visualize the Large Data 

plt.figure(figsize=(26,10))
sns.barplot(data=df, x='Age', y='Survived', hue='Sex',estimator=sum,errorbar=None)
plt.title('Barplot show casing Distribution between Age & Survived in accordance to Sex')
plt.xticks(rotation=45)

# Reduce font size
plt.xlabel('Age', fontsize=22)
plt.ylabel('Survived', fontsize=12)

# Limit the number of bars to show only the top N categories
top_n = 20  # Show top N values
top_values = df['Age'].value_counts().nlargest(top_n).index
filtered_data = df[df['Age'].isin(top_values)]
sns.barplot(data=filtered_data, x='Age', y='Survived')

# Show plot
plt.title('Top {} Categories'.format(top_n), fontsize=24)
plt.show()
plt.show()
```

Gender Distribution Across Age Categories:
The graph shows that, in most age categories, the percentage of male respondents is higher than that of female respondents.
This consistent trend suggests that males are more prevalent across various age groups within the dataset.
